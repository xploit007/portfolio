```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
data_file_path = 'Goal7_Transformed.csv'  # Update with your local path
electricity_access_data = pd.read_csv(data_file_path)
latest_year = 2020  # Assuming 2020 is the latest year in your dataset

# Extract necessary columns
latest_year_access_column = f'Access to electricity (% of population)_{latest_year}'
map_data = electricity_access_data[['Entity_Entity', latest_year_access_column]].copy()
map_data.columns = ['Country', 'Access']

# Load world geometry
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filter for only African countries
africa = world[world['continent'] == 'Africa']

# Merge African data with electricity access data
africa = africa.merge(map_data, how='left', left_on='name', right_on='Country')

# Define low access threshold
low_access_threshold = 50  # less than 50% access to electricity

# Plot
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
africa.boundary.plot(ax=ax, linewidth=1)
africa.plot(column='Access', ax=ax, legend=True, cmap='RdYlGn', legend_kwds={'label': "Access to Electricity (% of population)", 'orientation': "vertical"},
           missing_kwds={"color": "lightgray"})  # using RdYlGn color map for intuitive color coding

# Label countries with low access
for idx, row in africa.iterrows():
    if pd.notna(row['Access']) and row['Access'] < low_access_threshold:
        plt.annotate(text=row['name'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y),
                     xytext=(3, 3), textcoords="offset points", color='black')

ax.set_title('Highlighting Low Access Areas', fontsize=20)
ax.set_xlim([-25, 60])  # Adjust these values based on the extent you want to cover
ax.set_ylim([-40, 40])  # Adjust these values based on the extent you want to cover
ax.set_axis_off()
plt.show()
```

### Step 6: Forecasting Electricity Access
Use a SARIMAX model to forecast future electricity access rates. This involves fitting a model and predicting future values.

```python
from statsmodels.tsa.stattools import adfuller

# Create a time series data for electricity access
electricity_access_ts = pd.Series(data=average_electricity_access.values, index=pd.to_datetime(years_electricity, format='%Y'))

# Perform Augmented Dickey-Fuller test to check stationarity
adf_test = adfuller(electricity_access_ts)

# Display the ADF test results
adf_test_results = {
    'ADF Statistic': adf_test[0],
    'p-value': adf_test[1],
    'Critical Values': adf_test[4]
}

adf_test_results
```

```python
# Differencing the series once
electricity_access_ts_diff = electricity_access_ts.diff().dropna()

# Perform Augmented Dickey-Fuller test on the differenced series
adf_test_diff = adfuller(electricity_access_ts_diff)

# Display the ADF test results for the differenced series
adf_test_diff_results = {
    'ADF Statistic': adf_test_diff[0],
    'p-value': adf_test_diff[1],
    'Critical Values': adf_test_diff[4]
}

adf_test_diff_results
```

```python
# Manually specifying and fitting an ARIMA model
sarimax_model = SARIMAX(electricity_access_ts, order=(1, 2, 1), seasonal_order=(0, 0, 0, 0), enforce_stationarity=False, enforce_invertibility=False)
sarimax_model_fit = sarimax_model.fit(disp=False)

# Display the summary of the fitted model
sarimax_model_fit.summary()
```

### Step 7: Plot Forecast
Visualize the forecasted electricity access rates alongside historical data.

```python
# Forecasting from 2021 to 2030
forecast_years = pd.date_range(start='2021', periods=10, freq='AS')
forecast_results = sarimax_model_fit.get_forecast(steps=10)
forecast_mean = forecast_results.predicted_mean

# Confidence intervals for the forecasts
forecast_ci = forecast_results.conf_int()
forecast_years = pd.date_range(start='2021', periods=10, freq='AS-JAN')  # Ensure consistent datetime format

# Plot the historical data and forecasts without the confidence interval shading
plt.figure(figsize=(10, 6))
plt.plot(electricity_access_ts.index, electricity_access_ts, label='Historical Data', marker='o')
plt.plot(forecast_years, forecast_mean, label='Forecast', marker='o', linestyle='--', color='red')
plt.title('Forecast of Electricity Access Rates (2021-2030)')
plt.xlabel('Year')
plt.ylabel('Electricity Access (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# For the trend analysis, we'll focus on the latest available year (2020) and the earliest year in the dataset (2000)
# We'll calculate the growth of renewable energy share and production as well as fossil fuel and nuclear production

# Function to extract growth information between two years for a given energy type
def calculate_growth(data, energy_type, year_start, year_end):
    start_col = f'Electricity from {energy_type} (TWh)_{year_start}'
    end_col = f'Electricity from {energy_type} (TWh)_{year_end}'
    if start_col in data.columns and end_col in data.columns:
        data[f'Growth {energy_type}'] = (data[end_col] - data[start_col]) / data[start_col]
    return data[[f'Growth {energy_type}']].dropna()

# Calculate growth for renewables and fossil fuels between 2000 and 2020
growth_renewables = calculate_growth(goal7_data.copy(), 'renewables', '2000', '2020')
growth_fossil_fuels = calculate_growth(goal7_data.copy(), 'fossil fuels', '2000', '2020')

# Calculate the change in renewable energy share in total energy consumption
share_start_col = 'Renewable energy share in the total final energy consumption (%)_2000'
share_end_col = 'Renewable energy share in the total final energy consumption (%)_2020'
goal7_data['Change in Renewable Share'] = goal7_data[share_end_col] - goal7_data[share_start_col]

# Show summary statistics for growth and share change
growth_summary = {
    'Growth Renewables': growth_renewables['Growth renewables'].describe(),
    'Growth Fossil Fuels': growth_fossil_fuels['Growth fossil fuels'].describe(),
    'Change in Renewable Share': goal7_data['Change in Renewable Share'].describe()
}

growth_summary
```

```python
# Let's check if the columns related to the change in access are correctly calculated
# We will check if the relevant columns for electricity access in 2000 and 2020 are present
# Then we will compute the change in access correctly

# Check if the access columns are in the DataFrame
access_column_2000 = 'Access to electricity (% of population)_2000'
access_column_2020 = 'Access to electricity (% of population)_2020'

if access_column_2000 in goal7_data.columns and access_column_2020 in goal7_data.columns:
    # Compute the change in access between 2000 and 2020
    goal7_data['Change in Access 2000-2020'] = goal7_data[access_column_2020] - goal7_data[access_column_2000]

# Now we calculate the growth for renewables considering only the rows that have valid data for the year 2000
valid_start_values_renewables = (goal7_data['Electricity from renewables (TWh)_2000'] > 0) & (goal7_data['Electricity from renewables (TWh)_2000'].notna())

# Calculate the growth rate for renewables where valid
goal7_data.loc[valid_start_values_renewables, 'Growth renewables 2000-2020'] = (
    goal7_data.loc[valid_start_values_renewables, 'Electricity from renewables (TWh)_2020'] - 
    goal7_data.loc[valid_start_values_renewables, 'Electricity from renewables (TWh)_2000']
) / goal7_data.loc[valid_start_values_renewables, 'Electricity from renewables (TWh)_2000']

# Check if all necessary columns are present and then perform correlation
necessary_columns = [
    'Growth renewables 2000-2020', 
    'Change in Renewable Share', 
    'Change in Access 2000-2020'
]

if all(col in goal7_data.columns for col in necessary_columns):
    # Prepare data for correlation analysis
    correlation_data = goal7_data[necessary_columns].dropna()
    
    # Calculate correlation matrix
    correlation_matrix = correlation_data.corr()
else:
    # If any necessary columns are missing, we list which ones are not present
    missing_cols = [col for col in necessary_columns if col not in goal7_data.columns]
    correlation_matrix = f"Missing columns for correlation analysis: {missing_cols}"

correlation_matrix
```

```python
# For the correlation analysis, we will use the most recent year with a complete dataset (which appears to be 2020)
# We will correlate the growth in renewable energy production and the change in renewable energy share with changes in electricity access

# We need to identify the column for access to electricity in 2020
access_column_2020 = 'Access to electricity (% of population)_2020'
growth_column_renewables = 'Growth renewables'
change_share_column_renewables = 'Change in Renewable Share'

# We need the change in access to electricity from the baseline year 2000 to 2020
goal7_data['Change in Access 2000-2020'] = goal7_data[access_column_2020] - goal7_data['Access to electricity (% of population)_2000']

# We need to re-calculate the growth for renewables considering the starting and ending values
# We need to handle cases with zero or missing values to avoid infinite or undefined growth rates

# Filter out rows where the starting value is zero or missing to avoid division by zero or NaN issues
valid_start_values = (goal7_data['Electricity from renewables (TWh)_2000'] > 0)

# Calculate the growth rate where valid
goal7_data.loc[valid_start_values, 'Growth renewables 2000-2020'] = (
    goal7_data.loc[valid_start_values, 'Electricity from renewables (TWh)_2020'] - 
    goal7_data.loc[valid_start_values, 'Electricity from renewables (TWh)_2000']
) / goal7_data.loc[valid_start_values, 'Electricity from renewables (TWh)_2000']

# Now we can prepare our DataFrame for the correlation analysis
correlation_data = goal7_data.loc[
    valid_start_values,
    ['Growth renewables 2000-2020', 'Change in Renewable Share', 'Change in Access 2000-2020']
].dropna()

# Function to safely calculate growth to handle division by zero or missing data
def safe_growth_calculation(df, energy_type, start_year, end_year):
    start_col = f'Electricity from {energy_type} (TWh)_{start_year}'
    end_col = f'Electricity from {energy_type} (TWh)_{end_year}'
    
    # Filter out rows where the starting value is zero or missing to avoid division by zero or NaN issues
    valid_start = (df[start_col] > 0)
    
    # Calculate the growth rate where valid
    df.loc[valid_start, f'Growth {energy_type} {start_year}-{end_year}'] = (
        df.loc[valid_start, end_col] - df.loc[valid_start, start_col]
    ) / df.loc[valid_start, start_col]
    
    return df

# Apply growth calculation for fossil fuels and nuclear
goal7_data = safe_growth_calculation(goal7_data, 'fossil fuels', '2000', '2020')
goal7_data = safe_growth_calculation(goal7_data, 'nuclear', '2000', '2020')

# Prepare the data for correlation analysis
correlation_data = goal7_data.loc[
    valid_start_values,  # Ensure we only consider valid starting values for renewables
    [
        'Growth renewables 2000-2020', 
        'Growth fossil fuels 2000-2020',
        'Growth nuclear 2000-2020',
        'Change in Renewable Share', 
        'Change in Access 2000-2020'
    ]
].dropna()

# Calculating the correlation matrix again
correlation_matrix = correlation_data.corr()

correlation_matrix
```
